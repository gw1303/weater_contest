```python
df_dict = {}

for i in ['2016','2017','2018','2019','2020'] :
    
    main = pd.read_csv(f'동네예보/{i}_기온.csv').iloc[:,:3]
    
    main['hour'] = main['hour'].map(str)
    
    
    main = main[(main['forecast'] == 25) | (main['forecast'] == 46)]
    main['forecast'] = main['forecast'].map(str)
    
    main[' format: day'] = main[' format: day'].map(str.strip)
    
    main = main.reset_index(drop=True)
    
    year = []
    
    for j in range(len(main)) : 
        year.append(i)
        main['hour'][j] = main['hour'][j][:-4]
        main['forecast'][j] = main['forecast'][j][:-2]
        
    main['year'] = year
    res = {}
    
    for j in ['기온','습도'] :
        temp = pd.read_csv(f'동네예보/{i}_{j}.csv')
        temp = temp[(temp['forecast'] == 25) | (temp['forecast'] == 46)]
        res[j] = list(temp.iloc[:,3].values)
        
#     print(len(res['기온']),len(res['습도']),len(res['강수량']))
    df = pd.DataFrame({'temp':res['기온'], 'hum':res['습도']})
    
    main = main.join(df)
    
    main = main[['year',' format: day','hour','forecast','temp','hum']]
    main.columns =['year', 'day', 'hour', 'forecast', 'temp', 'hum']
    
    main.to_csv(f'{i}_통합.csv')
    
    df_dict.update({i:main})
```

```python
# 세계통합시간 한국시간으로 변경

for i in ['2016','2017','2018','2019','2020'] :
    df = pd.read_csv(f'동네예보_통합_수정(25,46)/{i}_통합.csv', index_col=0)
    
    df['datetime'] = None
    
    for j in range(len(df)) :
        temp_year = df['year'][j]
        temp_month = df['month'][j]
        temp_day = df['day'][j]
        temp_hour = df['hour'][j]
    
        add_time = datetime(temp_year,temp_month,temp_day,temp_hour) + timedelta(hours=10)
        add_time
        
        df['datetime'][j] = add_time
        
        df['year'][j] = add_time.year
        df['month'][j] = add_time.month
        df['day'][j] = add_time.day
        df['hour'][j] = add_time.hour
    
    df = df[['datetime','year', 'month', 'day', 'hour', 'forecast', 'temp', 'hum']]
    
    df.to_csv(f'{i}_시간수정(datetime).csv', index=False)
```

```python
weather['datetime_24'] = weather['datetime'].astype('datetime64') + timedelta(days=1)
weather['datetime_48'] = weather['datetime'].astype('datetime64') + timedelta(days=2)
weather

weather = weather[['datetime', 'datetime_24', 'temp_24', 'hum_24', 'datetime_48',
       'temp_48', 'hum_48']]

weather

weather['datetime'] = weather['datetime'].map(str.strip)
```

```python
weather.resample('10T').asfreq()
```

```python
#  날짜 데이터 1시간 단위로 쪼개기

weather.index = list(weather['datetime'].astype('datetime64'))
weather

weater_resample = weather.resample('10T').asfreq()
weater_resample['datetime'] = list(weater_resample.index)

# weater_resample = weater_resample.reset_index(drop=True)
# weater_resample

weater_resample['datetime_24'] = weater_resample['datetime'].astype('datetime64') + timedelta(days=1)
weater_resample['datetime_48'] = weater_resample['datetime'].astype('datetime64') + timedelta(days=2)

weater_resample['datetime'] = weater_resample['datetime'].astype('str')

weater_resample
```

```python
# 쪼갠 데이터 보간
weater_resample = weater_resample.interpolate(method='time')
weater_resample
```

```python
df['plant1_train.mea_ddhr'] = df['plant1_train.mea_ddhr'].astype('datetime64')
df2['plant2_train.mea_ddhr'] = df2['plant2_train.mea_ddhr'].astype('datetime64')

df['plant1_train.mea_ddhr'] = df['plant1_train.mea_ddhr'].astype('str')
df2['plant2_train.mea_ddhr'] = df2['plant2_train.mea_ddhr'].astype('str')

df['plant1_train.mea_ddhr'] = df['plant1_train.mea_ddhr'].map(str.strip)
df2['plant2_train.mea_ddhr'] = df2['plant2_train.mea_ddhr'].map(str.strip)

# 공장 데이터에 날씨 데이터 병합
df = df.merge(weater_resample, left_on='plant1_train.mea_ddhr', right_on='datetime', how='outer')
df2 = df2.merge(weater_resample, left_on='plant2_train.mea_ddhr', right_on='datetime', how='outer')

# 공장데이터가 있는 날짜만 출력
df = df[df['plant1_train.mea_ddhr'].notnull()]
df2 = df2[df2['plant2_train.mea_ddhr'].notnull()]

df.to_csv('plant1_train_merge.csv', index=False)
df2.to_csv('plant2_train_merge.csv', index=False)
```
